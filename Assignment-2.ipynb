```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/House Price India.csv')
print(data.head())
```

```python
# Histogram
sns.histplot(data['living area'])
plt.show()
```

```python
# Bar chart
sns.countplot(x='number of bedrooms', data=data)
plt.show()
```

```python
# Scatter plot
sns.scatterplot(x='number of bedrooms', y='number of bathrooms', data=data)
plt.show()
```

```python
# Box plot or Violin plot
sns.boxplot(x='waterfront present', y='lot area', data=data)
plt.show()
```

```python
# Pairplot for multiple numerical columns
sns.pairplot(data[['living area', 'lot area', 'Price']])
plt.show()

# Correlation heatmap
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
# This will give you count, mean, std deviation, min, 25%, 50%, 75% and max values for numerical columns
desc_stats = data.describe()
print(desc_stats)

# For categorical columns
cat_desc = data['condition of the house'].value_counts()
print(cat_desc)
```

```python
missing_values = data.isnull().sum()
print(missing_values)
```
