```python
import numpy as np
import pandas as pd
```

```python
# Using numpy to generate a random array of shape (4,4)
data = np.random.rand(4, 4)

# Creating a DataFrame using the random data
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3', 'Feature4'])
```

```python
print(df)
```

```python
#Renaming the task
df.columns = ['Random value 1', 'Random value 2', 'Random value 3', 'Random value 4']
```

```python
print(df)
```

```python
statistics = df.describe()
print(statistics)
```

```python
# To check the null values
null_values = df.isnull().sum()
print("Null values in each column:\n", null_values)
```

```python
column_data_types = df.dtypes
print("\nData type of each column:\n", column_data_types)
```

```python
#Location Method
selected_columns_loc = df.loc[:, ['Random value 2', 'Random value 3']]
print("Using .loc:\n", selected_columns_loc)
```

```python
#Index location Method
selected_columns_iloc = df.iloc[:, [1, 2]]
print("\nUsing .iloc:\n", selected_columns_iloc)
```
